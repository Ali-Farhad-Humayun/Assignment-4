#1. List out the features of HTML 5.

Answer: Here are the key features of HTML5 in brief:

Semantic Elements: Introduces meaningful tags like <header>, <footer>, <article>, <section>, <nav>, and <aside> for better document structure and readability.

Multimedia Support: Native support for audio (<audio>) and video (<video>) embedding without third-party plugins like Flash.

Canvas Element: The <canvas> tag allows for dynamic, scriptable rendering of 2D graphics and animations.

Form Enhancements: New input types (date, email, url, range, etc.) and attributes (placeholder, required, autofocus) improve user experience and validation.

Offline and Storage:

Local Storage and Session Storage for client-side data storage.

Geolocation API: Allows websites to access a user's geographical location (with permission).

WebSockets: Enables real-time, two-way communication between the client and server.

Responsive Images: Features like srcset and <picture> allow for better image optimization across devices.

Accessibility Improvements: Enhanced support for ARIA (Accessible Rich Internet Applications) attributes and semantic elements for better accessibility.

Cross-Browser Compatibility: Designed to work consistently across modern browsers.

Doctype Simplification: Simplified doctype declaration (<!DOCTYPE html>) and character encoding (<meta charset="UTF-8">).

Microdata: Allows embedding machine-readable data in HTML for better SEO and data interpretation.

HTML5 revolutionized web development by making it more interactive, dynamic, and user-friendly.

#2. What are HTML Entities? List out 5 commonly used HTML entities.

Answer: HTML entities are special codes used to represent characters that have reserved meanings in HTML or characters that cannot be typed directly (e.g., symbols, special characters, or non-ASCII characters). They start with an ampersand (&) and end with a semicolon (;).

Commonly Used HTML Entities:
Here are 5 commonly used HTML entities:

&lt;

Represents the less-than sign (<).

Used to display < in HTML without being interpreted as the start of a tag.

&gt;

Represents the greater-than sign (>).

Used to display > in HTML without being interpreted as the end of a tag.

&amp;

Represents the ampersand (&).

Used to display & in HTML without being interpreted as the start of an entity.

&nbsp;

Represents a non-breaking space.

Used to create a space that prevents line breaks.

&copy;

Represents the copyright symbol (Â©).

Used to display the copyright symbol in text.

#3. Define accessibility in the context of web development. Discuss why it's essential to create accessible websites and how it benefits different user group.

Answer: Accessibility in Web Development:
Accessibility refers to designing and developing websites and applications that are usable by everyone, including people with disabilities (e.g., visual, auditory, motor, or cognitive impairments). It ensures equal access to information and functionality.

Why Accessibility is Essential:
Inclusivity: Ensures everyone, regardless of ability, can access and interact with web content.

Legal Compliance: Many countries have laws (e.g., ADA, WCAG) requiring accessible websites.

Improved SEO: Accessible practices (e.g., semantic HTML, alt text) often align with search engine optimization.

Better User Experience: Enhances usability for all users, including those using mobile devices or slow internet.

Broader Audience Reach: Expands your audience by accommodating diverse user needs.

Benefits for Different User Groups:
Visually Impaired: Screen readers and proper contrast help navigate content.

Hearing Impaired: Captions and transcripts make audio/video content accessible.

Motor Impaired: Keyboard navigation and voice commands enable interaction.

Cognitive Impairments: Clear layouts and simple language improve understanding.

Elderly Users: Larger text and intuitive designs enhance usability.

By prioritizing accessibility, developers create a more inclusive and user-friendly web.

#4. List any 3 ways which help us in improving the accessibility of HTML.

Answer: Here are 3 ways to improve the accessibility of HTML:

Use Semantic HTML:

Replace non-semantic tags (e.g., <div>, <span>) with semantic elements like <header>, <nav>, <main>, <article>, and <footer> to provide better structure and context for screen readers.

Add Alt Text to Images:

Use the alt attribute in <img> tags to describe images for visually impaired users and when images fail to load.

Example: <img src="logo.png" alt="Company Logo">.

Ensure Keyboard Accessibility:

Make all interactive elements (e.g., buttons, links, forms) navigable using the keyboard.

Use tabindex to manage focus order and ensure proper focus indicators.

These practices enhance usability for all users, including those with disabilities.
